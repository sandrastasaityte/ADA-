```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "ADA.csv"
df = pd.read_csv(file_path) 

# Line plot of 'close' prices over time
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['close'])
plt.title('Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.show()

# Scatter plot of 'high' prices vs 'low' prices
plt.figure(figsize=(8, 6))
sns.scatterplot(x='high', y='low', data=df)
plt.title('High Prices vs Low Prices')
plt.xlabel('High Price')
plt.ylabel('Low Price')
plt.show()

# Histogram of 'Volume ADA'
plt.figure(figsize=(8, 6))
plt.hist(df['Volume ADA'], bins=20, edgecolor='black')
plt.title('Distribution of Volume ADA')
plt.xlabel('Volume ADA')
plt.ylabel('Frequency')
plt.show()

# Box plot of 'tradecount'
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['tradecount'])
plt.title('Boxplot of Trade Count')
plt.xlabel('Trade Count')
plt.show()
```

## Better version Date vs Moth/Year 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "ADA.csv"
df = pd.read_csv(file_path) 

# Extract the date column
df['date'] = pd.to_datetime(df['date'])

# Extract month and year
df['month_year'] = df['date'].dt.strftime('%b %Y')

# Line plot of 'close' prices over time with month/year on x-axis
plt.figure(figsize=(10, 6))
plt.plot(df['month_year'], df['close'])
plt.title('Closing Prices Over Time')
plt.xlabel('Month/Year')
plt.ylabel('Closing Price')
plt.xticks(rotation=45, ha='right')  # Adjust rotation and alignment for better visibility
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "ADA.csv"
df = pd.read_csv(file_path) 

# Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Extract month and year
df['month_year'] = df['date'].dt.strftime('%b %Y')

# Box plot of 'close' prices by month/year
plt.figure(figsize=(12, 8))
sns.boxplot(x='month_year', y='close', data=df)
plt.title('Boxplot of Closing Prices by Month/Year')
plt.xlabel('Month/Year')
plt.ylabel('Closing Price')
plt.xticks(rotation=45, ha='right')  # Adjust rotation and alignment for better visibility
plt.show()
```

```python
# Access directly the Database ONLY FOR STUDY PURPOSE.  

import pandas as pd
import pyodbc

# Replace these with your MS-SQL Server connection details
server = 'your_server'
database = 'your_database'
username = 'your_username'
password = 'your_password'

# Create an MS-SQL Server connection string
connection_string = f'DRIVER={{ODBC Driver 17 for SQL Server}};SERVER={server};DATABASE={database};UID={username};PWD={password}'

# Create an MS-SQL Server connection
connection = pyodbc.connect(connection_string)

# Define the stored procedure (function) name
stored_procedure_name = 'your_stored_procedure_name'

# Parameters for the stored procedure
parameter1 = 'value1'
parameter2 = 'value2'

# SQL query to call the stored procedure
sql_query = f"EXEC {stored_procedure_name} @param1='{parameter1}', @param2='{parameter2}'"

# Execute the SQL query and read the result into a Pandas DataFrame
df = pd.read_sql_query(sql_query, connection)

# Close the MS-SQL Server connection
connection.close()

# Print the Pandas DataFrame
print("Stored Procedure Result:")
print(df)
```
